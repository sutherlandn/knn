````python
#1
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
 print(y_test[i],'\n') # Print the label
 print(X_test[i],'\n') # Print the matrix of values
 plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
 plt.show()
````

````python
#2
print(f'X training shape:{X_train.shape}')
print(f'X test shape:{X_test.shape}')
print(f'X training shape at 0: {X_train[0].shape}')
print(f'X testing shape at 0: {X_test[0].shape}')
print(f'Y training shape{y_train.shape}')
print(f'Y testing shape{y_test.shape}')
````

````python
#3
import numpy as np
import pandas as pd

Z_train = X_train.reshape(X_train.shape[0], -1)  # Flatten each 28x28 image into a 784-length vector

Z_test = X_test.reshape(X_test.shape[0], -1)  # Flatten each 28x28 image into a 784-length vector

# Convert the reshaped data into pandas DataFrames
Z_train_df = pd.DataFrame(Z_train)
Z_test_df = pd.DataFrame(Z_test)

# Optionally, save the reshaped data to CSV
Z_train_df.to_csv('./data/Z_train.csv', index=False)
Z_test_df.to_csv('./data/Z_test.csv', index=False)
````

````python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Determine the optimal k:
k_bar = 20
accuracy = []

for k in range(k_bar):
    print('Current k: ', str(k+1))
    knn = KNeighborsClassifier(n_neighbors=k+1)  # Create classifier instance
    predictor = knn.fit(Z_train, y_train)  # Fit model with NumPy arrays
    accuracy.append(predictor.score(Z_test, y_test))  # Store accuracy

accuracy_max = np.max(accuracy)  # highest recorded accuracy
max_index = np.where(accuracy == accuracy_max)[0]
k_star = max_index + 1  # Find the optimal value of k
print(k_star)

plt.plot(np.arange(1, k_bar + 1), accuracy)  # Plot accuracy by k
plt.xlabel("k")
plt.title("Optimal k: " + str(k_star))
plt.ylabel('Accuracy')
plt.show()
# Example code was wonky, used Chat here to clean some stuff up^
````

````python
#4
knn = KNeighborsClassifier(n_neighbors=3)
predictor = knn.fit(Z_train, y_train)
y_hat = predictor.predict(Z_test)

# accuracy score
accuracy = knn.score(Z_test, y_test)
print('Accuracy: ', accuracy)

# confusion matrix
conf_matrix = pd.crosstab(y_test, y_hat)
print(conf_matrix)
````

With k = 3, we got 97% accuracy (higher than in the example solutions!)
